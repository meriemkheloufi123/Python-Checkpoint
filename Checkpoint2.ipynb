1-Write a program which will find all such numbers which are divisible by 7 but are not a multiple of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a list.

```
Mylist=[]
print("Les nombres inclus entre 2000 et 3200 et qui sont divisibles par 7 et non multiples de 5 sont : ")
for i in range(2000, 3201):
    if i%7 == 0 and i%5 != 0 :
        Mylist.append(i)  #append c pour ajouter des élémets à la liste
print(Mylist)
```

2- Write a program which can compute the factorial of a given number.

```
n = int(input(' entrez un nombre: '))
fact = 1
for i in range(1, n+1) :
    fact = i * fact 
print(n, '! = ' ,fact)
```

3- With a given integer number n, write a program to generate a dictionary that contains (i, i*i) such that is an integral number between 1 and n (both included). and then the program should print the dictionary. 

```
n = int(input('entrez un nombre: '))
Mydict={}
for i in range(1,n+1):
    Mydict[i]=i*i
print(Mydict)
```

4- Given a non-empty string and an int n, return a new string where the char at index n has been removed. The value of n will be a valid index of a char in the original string

```
CH = str(input("donnez un mot : "))
n = int(input("donnez un indice :"))
def remove_char(str, n):
      debut = str[:n] 
      fin = str[n+1:]
      return debut + fin
print(remove_char(CH, n))
```

5- Write a NumPy program to convert a NumPy array into a Python list structure.

6- Write a NumPy program to compute the covariance matrix of two given arrays.

```
import numpy as np
newarray = np.array([0,1,2]) 
print("Original array 1 :")
print(newarray)
array = np.array([2,1,0]) 
print("Original array 2 :")
print(array)
print("The covarinace matrix:", np.cov(newarray,array))
```

7- Write a program that calculates and prints the value according to the given formula: Q = Square root of [(2 * C * D)/H]

```
import math
C=50
H=30
D = []
print( "Donnez les valeurs de D : ")
items=[x for x in input().split(',')]
print("Les racines carrées de D : ")
for d in items:
    D.append(str(int(round(math.sqrt(2*C*float(d)/H)))))
print(','.join(D))
```
